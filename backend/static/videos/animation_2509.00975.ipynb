# Self-Exploring Language Models for Explainable Link Forecasting on Temporal Graphs via Reinforcement Learning

Forecasting future links is a central task in temporal graph (TG) reasoning, requiring models to leverage historical interactions to predict upcoming ones. Traditional neural approaches, such as temporal graph neural networks, achieve strong performance but lack explainability and cannot be applied to unseen graphs without retraining. Recent studies have begun to explore using large language models (LLMs) for graph reasoning, but most of them are constrained to static graphs or small synthetic TGs and lack the evaluation of the quality of reasoning traces generated by LLMs. In this work, we present Reasoning-Enhanced Learning for Temporal Graphs (ReaL-TG), a reinforcement learning framework that fine-tunes LLMs to perform explainable link forecasting on real-world TGs. ReaL-TG uses outcome-based reward to encourage models to self-explore reasoning strategies from graph structure and to produce explanations that directly justify their predictions. To enable evaluation on LLM-generated reasoning traces, we propose a new evaluation protocol combining ranking metrics with an LLM-as-a-Judge system that assesses both the quality of reasoning and the impact of hallucinations. Experiments with ReaL-TG-4B, obtained by fine-tuning Qwen3-4B under our framework, show that it outperforms much larger frontier LLMs, including GPT-5 mini, on ranking metrics, while producing high-quality explanations confirmed by both the LLM judge and human evaluation.

```python
%%writefile animation.py
from manim import *

class PaperAnimation(Scene):
    def construct(self):
        # 1. Title and Abstract Intro
        title_text = Text(
            "Self-Exploring Language Models for Explainable Link Forecasting on Temporal Graphs via Reinforcement Learning",
            font_size=48,
            weight=BOLD,
            line_spacing=1.5
        ).to_edge(UP)
        
        real_tg_label = Text("ReaL-TG Framework", font_size=36, color=BLUE).next_to(title_text, DOWN, buff=0.5)

        self.play(Write(title_text))
        self.play(Write(real_tg_label))
        self.wait(1)
        self.play(FadeOut(title_text), FadeOut(real_tg_label))

        # 2. Temporal Graph Concept
        tg_title = Text("Temporal Graphs (TG)").to_edge(UP).set_color(ORANGE)
        self.play(Write(tg_title))

        # Nodes
        node_a = Circle(radius=0.5, color=BLUE, fill_opacity=0.8).shift(LEFT * 3 + UP * 1)
        node_b = Circle(radius=0.5, color=BLUE, fill_opacity=0.8).shift(RIGHT * 3 + UP * 1)
        node_c = Circle(radius=0.5, color=BLUE, fill_opacity=0.8).shift(LEFT * 1 + DOWN * 2)
        node_d = Circle(radius=0.5, color=BLUE, fill_opacity=0.8).shift(RIGHT * 1 + DOWN * 2)

        label_a = Text("A", font_size=30).move_to(node_a)
        label_b = Text("B", font_size=30).move_to(node_b)
        label_c = Text("C", font_size=30).move_to(node_c)
        label_d = Text("D", font_size=30).move_to(node_d)

        nodes = VGroup(node_a, node_b, node_c, node_d, label_a, label_b, label_c, label_d)
        self.play(Create(nodes))

        # Edges (Time 1)
        edge_ab_t1 = Line(node_a.get_center(), node_b.get_center(), color=GREEN)
        edge_ac_t1 = Line(node_a.get_center(), node_c.get_center(), color=GREEN)
        
        # Create edges individually for better management of the graph_elements group later
        self.play(Create(edge_ab_t1), Create(edge_ac_t1))
        self.wait(0.5)

        # Edges (Time 2) - new edge, old one fades
        edge_bd_t2 = Line(node_b.get_center(), node_d.get_center(), color=GREEN)
        self.play(FadeOut(edge_ac_t1), Create(edge_bd_t2)) # edge_ac_t1 fades out here
        self.wait(0.5)

        # Edges (Time 3) - another new edge
        edge_cd_t3 = Line(node_c.get_center(), node_d.get_center(), color=GREEN)
        self.play(Create(edge_cd_t3))
        self.wait(1)

        # Group currently visible graph elements for fading out (node group, and the active edges)
        graph_elements = VGroup(nodes, edge_ab_t1, edge_bd_t2, edge_cd_t3) # Changed to VGroup
        self.play(FadeOut(graph_elements), FadeOut(tg_title))
        self.wait(0.5)

        # 3. LLM and Link Forecasting
        llm_title = Text("Large Language Models (LLM)").to_edge(UP).set_color(PURPLE)
        self.play(Write(llm_title))

        llm_box = Rectangle(width=4, height=2, color=PURPLE, fill_opacity=0.2).center()
        llm_text = Text("LLM", font_size=40, color=PURPLE).move_to(llm_box.get_center())
        llm_group = VGroup(llm_box, llm_text) # Changed to VGroup
        self.play(Create(llm_group))
        self.wait(1)

        # Bring back simplified graph
        node_1 = Circle(radius=0.4, color=BLUE, fill_opacity=0.8).shift(LEFT * 4 + UP * 1.5)
        node_2 = Circle(radius=0.4, color=BLUE, fill_opacity=0.8).shift(LEFT * 2 + UP * 1.5)
        node_3 = Circle(radius=0.4, color=BLUE, fill_opacity=0.8).shift(LEFT * 4 + DOWN * 0.5)
        node_4 = Circle(radius=0.4, color=BLUE, fill_opacity=0.8).shift(LEFT * 2 + DOWN * 0.5)
        
        label_1 = Text("1", font_size=24).move_to(node_1)
        label_2 = Text("2", font_size=24).move_to(node_2)
        label_3 = Text("3", font_size=24).move_to(node_3)
        label_4 = Text("4", font_size=24).move_to(node_4)

        current_edge_12 = Line(node_1.get_center(), node_2.get_center(), color=GREEN)
        current_edge_13 = Line(node_1.get_center(), node_3.get_center(), color=GREEN)
        current_edge_34 = Line(node_3.get_center(), node_4.get_center(), color=GREEN)

        simplified_graph = VGroup(node_1, node_2, node_3, node_4, label_1, label_2, label_3, label_4, 
                                 current_edge_12, current_edge_13, current_edge_34) # Changed to VGroup
        
        self.play(Create(simplified_graph))
        self.wait(0.5)

        # Link Forecasting
        input_arrow = Arrow(simplified_graph.get_right(), llm_box.get_left(), buff=0.1, color=WHITE)
        input_text = Text("Graph History", font_size=24, color=WHITE).next_to(input_arrow, UP)
        self.play(Create(input_arrow), Write(input_text))
        self.wait(0.5)

        prediction_text_label = Text("Prediction:", font_size=24, color=RED).next_to(llm_group, RIGHT, buff=0.8)
        self.play(Write(prediction_text_label))

        predicted_link_line = DashedLine(node_2.get_center(), node_4.get_center(), color=RED)
        predicted_link_text = Text("Future Link (2-4)", font_size=24, color=RED).next_to(predicted_link_line, UP)
        
        self.play(Create(predicted_link_line), Write(predicted_link_text))
        self.wait(1)

        link_forecast_group = VGroup(input_arrow, input_text, prediction_text_label, predicted_link_line, predicted_link_text) # Changed to VGroup
        self.play(FadeOut(llm_title), FadeOut(simplified_graph), FadeOut(llm_group), FadeOut(link_forecast_group))
        self.wait(0.5)

        # 4. Reinforcement Learning for Explainable Forecasting (ReaL-TG)
        real_tg_full_title = Text("ReaL-TG: Reinforcement Learning for Explainable Forecasting").to_edge(UP).set_color(BLUE)
        self.play(Write(real_tg_full_title))

        # LLM in center
        llm_box_center = Rectangle(width=3, height=1.5, color=PURPLE, fill_opacity=0.2).center()
        llm_text_center = Text("LLM", font_size=36, color=PURPLE).move_to(llm_box_center.get_center())
        llm_group_center = VGroup(llm_box_center, llm_text_center) # Changed to VGroup
        self.play(Create(llm_group_center))

        # Temporal Graph representation
        tg_rep = Text("Temporal Graph", font_size=28, color=ORANGE).shift(LEFT * 4 + UP * 2)
        tg_rep_box = Rectangle(width=3.5, height=1, color=ORANGE, fill_opacity=0.1).move_to(tg_rep)
        self.play(Create(tg_rep_box), Write(tg_rep))

        # Reinforcement Learning Module
        rl_module = Text("RL Module", font_size=28, color=GREEN).shift(LEFT * 4 + DOWN * 2)
        rl_module_box = Rectangle(width=3, height=1, color=GREEN, fill_opacity=0.1).move_to(rl_module)
        self.play(Create(rl_module_box), Write(rl_module))

        # Explainability Module / Evaluation
        exp_eval_module = Text("Explanation & Eval", font_size=28, color=YELLOW).shift(RIGHT * 4 + DOWN * 2)
        exp_eval_module_box = Rectangle(width=4.5, height=1, color=YELLOW, fill_opacity=0.1).move_to(exp_eval_module)
        self.play(Create(exp_eval_module_box), Write(exp_eval_module))
        
        # Connections
        # TG -> LLM (Observation)
        arrow_tg_llm = Arrow(tg_rep_box.get_right(), llm_box_center.get_left() + UP * 0.5, color=WHITE)
        label_tg_llm = Text("Observe Graph", font_size=20).next_to(arrow_tg_llm, UP, buff=0.1)
        self.play(Create(arrow_tg_llm), Write(label_tg_llm))

        # LLM -> Prediction & Explanation (Action)
        prediction_action_pos = llm_box_center.get_right() + UP * 0.5
        explanation_action_pos = llm_box_center.get_right() + DOWN * 0.5

        arrow_llm_pred = Arrow(llm_box_center.get_right(), prediction_action_pos + RIGHT * 1, color=RED)
        label_llm_pred = Text("Link Prediction", font_size=20, color=RED).next_to(arrow_llm_pred, UP)
        self.play(Create(arrow_llm_pred), Write(label_llm_pred))
        
        arrow_llm_exp = Arrow(llm_box_center.get_right(), explanation_action_pos + RIGHT * 1, color=YELLOW)
        label_llm_exp = Text("Explanation", font_size=20, color=YELLOW).next_to(arrow_llm_exp, DOWN)
        self.play(Create(arrow_llm_exp), Write(label_llm_exp))

        # Prediction -> RL Module (Outcome)
        arrow_pred_rl = Arrow(prediction_action_pos + RIGHT * 1, rl_module_box.get_top() + LEFT * 0.5, color=GREEN)
        label_pred_rl = Text("Outcome", font_size=20).next_to(arrow_pred_rl, RIGHT, buff=0.1)
        self.play(Create(arrow_pred_rl), Write(label_pred_rl))
        
        # Explanation -> Explanability Module (Evaluation)
        arrow_exp_eval = Arrow(explanation_action_pos + RIGHT * 1, exp_eval_module_box.get_top() + RIGHT * 0.5, color=YELLOW)
        label_exp_eval = Text("Trace Eval", font_size=20).next_to(arrow_exp_eval, LEFT, buff=0.1)
        self.play(Create(arrow_exp_eval), Write(label_exp_eval))

        # RL Module -> LLM (Reward)
        arrow_rl_llm = Arrow(rl_module_box.get_right(), llm_box_center.get_left() + DOWN * 0.5, color=WHITE)
        label_rl_llm = Text("Reward", font_size=20).next_to(arrow_rl_llm, DOWN, buff=0.1)
        self.play(Create(arrow_rl_llm), Write(label_rl_llm))

        # Combined feedback from RL and Explanability to LLM (Implicit: fine-tuning)
        # Let's add a conceptual arrow from explanation eval back to RL or LLM
        arrow_eval_rl = Arrow(exp_eval_module_box.get_left(), rl_module_box.get_bottom() + RIGHT * 0.5, color=WHITE)
        label_eval_rl = Text("Eval Feedback", font_size=20).next_to(arrow_eval_rl, DOWN, buff=0.1)
        self.play(Create(arrow_eval_rl), Write(label_eval_rl))

        self.wait(2)

        # Fade out everything at the end
        all_objects = Group(*self.mobjects) # Fix: Changed VGroup to Group to handle all Mobject types
        self.play(FadeOut(all_objects))
        self.wait(1)
```
